```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score,mean_squared_error,mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
```

```python
data = pd.read_csv('/Users/Ethan/Downloads/homelessdata.csv',encoding='latin1', low_memory=False)
```

```python
data['Enrollments Project Start Date']
```

```python
#changed the dates from an object to datetime
data['Enrollments Project Start Date'] = pd.to_datetime(data['Enrollments Project Start Date'], errors = 'coerce')
data['Enrollments Project Start Date']
```

```python
data.head(5)
```

```python
#dropped columns that were repetitive and not needed for conducting the prediction model
data.drop(columns = ['Clients Ethnicity','Clients American Indian or Alaska Native'
                    ,'Clients Asian','Clients Black or African American','Clients Native Hawaiian or Other Pacific Islander'
                    ,'Clients White','Entry Screen Education Degree','Entry Screen Employment Seeking',
                    'Update/Exit Custom Current Program Phase','Update/Exit Screen Housing Status'
                    ,'Enrollments Days from Project Start to Move-In',
                    'Enrollments Project Exit Date','Inbound Recidivism Days Since Last Exit',
                    'Outbound Recidivism Days Until Next Entry','Outbound Recidivism Next Entry Date',
                    'Program Inventory End Date','Program Inventory Start Date','Programs Added Date','Update/Exit Screen Exit Destination'
                    ,'Update/Exit Screen Housed on Exit','Clients Global Household - Head of Household (HoH)',
                    'Clients Discharge Status','Enrollments Last Exit Destination','Entry Screen Currently Fleeing Domestic Violence',
                    'Enrollments Average Days from Project Start to Move-In','Enrollments Average Days in Project',
                    'Entry Screen Non-Cash Benefit from Any Source','Entry Screen Income from any Source',
                    'Entry Screen Earned Income Tiers','Enrollments Count Children','Clients Unique Identifier'], inplace = True)
```

```python
data.dropna(inplace = True)
```

```python
data.describe()
```

```python
#only wanted data that were within the 10 year period of recording
#any enrollments that started before 2008 were removed

data1 = data.loc[data['Enrollments Project Start Date'] >= '1/1/2008']
data1
```

```python
#I only looked for enrollment days greater than 2 because if a client were to stay for less than 2 days
#it would not be helpful to predict short stays
data2 = data1.loc[data1['Enrollments Days in Project'] >= 2, :]
data2.describe()
```

```python
data3 = pd.get_dummies(data2, drop_first = True)
#changing categorical columns into dummies
```

```python
data3.head(10)
```

```python
#dropping this column because it is not needed in the prediction model
data3.drop(columns = ['Enrollments Project Start Date'], inplace = True)
```

```python
#dropping this column because this will have an impact on the predicted number of days
x = data3.drop('Enrollments Days in Project', axis = 1)
```

```python
#creating the x test, x train, y test, y train, doing a 20-80 split
labels = data3['Enrollments Days in Project']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,labels, test_size = 0.2)
```

```python
np.mean(y_train)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred_lr = lr.predict(x_test)
score_lr = lr.score(x_train,y_train)
print(score_lr)
print(lr.coef_[0])
print(lr.intercept_)
preco_lr = data2['Enrollments Days in Project']
predict_lr = pd.Series(pred_lr)
mae_lr = mean_absolute_error(y_test,pred_lr)
mse_lr = mean_squared_error(y_test,pred_lr)
rmse_lr = np.sqrt(mse_lr)

print('Mae_lr: ',mae_lr)
print('Mse_lr: ',mse_lr)
print('Rmse_lr: ',rmse_lr)
```

```python
#decision tree regressor
dtr = DecisionTreeRegressor(random_state=0)
dtr.fit(x_train,y_train)
pred_dtr = dtr.predict(x_test)
score_dtr = dtr.score(x_train,y_train)
print(score_dtr)
mae_dtr = mean_absolute_error(y_test,pred_dtr)
mse_dtr = mean_squared_error(y_test,pred_dtr)
rmse_dtr = np.sqrt(mse_dtr)
print('Mae_dtr: ',mae_dtr)
print('Mse_dtr: ',mse_dtr)
print('Rmse_dtr: ',rmse_dtr)
```

```python
#random forest
rfr = RandomForestRegressor(n_estimators=200)
rfr.fit(x_train,y_train)
score_rfr = rfr.score(x_train,y_train)
print(score_rfr)
pred_rfr = rfr.predict(x_test)
mae_rfr = mean_absolute_error(y_test,pred_rfr)
mse_rfr = mean_squared_error(y_test,pred_rfr)
rmse_rfr = np.sqrt(mean_squared_error(y_test,pred_rfr))
print('Mae: ',mae_rfr)
print('Mse: ',mse_rfr)
print('Rmse: ',rmse_rfr)
```

```python
#knn regressor
knr = KNeighborsRegressor(n_neighbors = 10)
knr.fit(x_train,y_train)
score_knr = knr.score(x_train,y_train)
print(score_knr)
pred_knr = knr.predict(x_test)
mae_knr = mean_absolute_error(y_test,pred_knr)
mse_knr = mean_squared_error(y_test,pred_knr)
rmse_knr = np.sqrt(mse_knr)
print('Mae_knr: ',mae_knr)
print('Mse_knr: ',mse_knr)
print('Rmse_knr: ',rmse_knr)
```

```python
#prediction of 411
def mape(actual,pred):
    return np.mean(np.abs((actual - pred) / actual)) * 100
actual = y_test
pred_set = [411]*12619
#Calculate the MAPE
result = mape(actual,pred_set)
#print the result
print("The mean absolute percentage error: ",result)
mae_knr = mean_absolute_error(y_test,pred_set)
mse_knr = mean_squared_error(y_test,pred_set)
rmse_knr = np.sqrt(mse_knr)
print('Mae_knr: ',mae_knr)
print('Mse_knr: ',mse_knr)
print('Rmse_knr: ',rmse_knr)
```
